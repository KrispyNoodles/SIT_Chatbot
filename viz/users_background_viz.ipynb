```python
file_path = '/home/krispy_noodles/SIT_Chatbot/assets/SIT User Study.xlsx'
```

```python
def user_comparisons():
    """
    Function to plot comparisons between Chatbot and Manual Search metrics from an Excel sheet,
    including average lines on the line graph and bar graph, using SIT-inspired colors.

    Parameters:
    - file_path: str, path to the Excel file.
    """
    import pandas as pd
    import matplotlib.pyplot as plt

    # SIT-inspired color palette
    sit_colors = ['#FF1F1F', '#000000', '#FFFFFF', '#F0F0F0', '#808080']

    # Load the sheet
    df = pd.read_excel(file_path, sheet_name='Feedback')
    
    # Clean the DataFrame
    df_cleaned = df.dropna().reset_index(drop=True)
    
    # Standardize column names to avoid formatting issues
    df_cleaned.columns = df_cleaned.columns.str.strip()
    
    # Count unique values for both categories
    sit_students = df_cleaned["SIT Students?"].value_counts()
    visited_counts = df_cleaned["Have they visited SIT's website before?"].value_counts()
    chatgpt_counts = df_cleaned["Have they used chatgpt before?"].value_counts()

    # Plot pie chart for 'SIT Students'
    plt.figure(figsize=(6, 6))
    plt.pie(sit_students, labels=sit_students.index, autopct='%1.1f%%', colors=[sit_colors[0], sit_colors[4]])
    plt.title("Are they SIT Students?", fontweight="bold")
    plt.show()

    # Plot pie chart for 'Visited SIT Website'
    plt.figure(figsize=(6, 6))
    plt.pie(visited_counts, labels=visited_counts.index, autopct='%1.1f%%', colors=[sit_colors[0], sit_colors[4]])
    plt.title("Have They Visited SIT's Website Before?", fontweight="bold")
    plt.show()

    # Plot pie chart for 'Used ChatGPT Before'
    plt.figure(figsize=(6, 6))
    plt.pie(chatgpt_counts, labels=chatgpt_counts.index, autopct='%1.1f%%', colors=[sit_colors[0], sit_colors[4]])
    plt.title("Have They Used ChatGPT Before?", fontweight="bold")
    plt.show()
```

```python
user_comparisons()
```
